```python
import os
import sys
import argparse
import cv2
import math
import numpy as np
from scipy import fftpack
from tqdm import tqdm
import glob
import random
import time
from collections import deque, defaultdict

# Optional deep learning parts
try:
    import torch
    import torch.nn as nn
    import torchvision
    from torchvision import transforms, models, datasets
    from torch.utils.data import DataLoader
    TORCH_AVAILABLE = True
except Exception:
    TORCH_AVAILABLE = False

# ---------------------------
# Utility: frame extraction
# ---------------------------
def extract_frames(video_path, every_n_frames=5, max_frames=None):
    """
    Extract frames from video_path.
    - every_n_frames: sample every Nth frame to reduce compute.
    - max_frames: cap total frames extracted.
    Returns list of (frame_index, BGR_image).
    """
    cap = cv2.VideoCapture(video_path)
    if not cap.isOpened():
        raise RuntimeError(f"Cannot open video: {video_path}")
    frames = []
    idx = 0
    out_count = 0
    fps = cap.get(cv2.CAP_PROP_FPS) or 30
    while True:
        ret, frame = cap.read()
        if not ret:
            break
        if idx % every_n_frames == 0:
            frames.append((idx, frame.copy()))
            out_count += 1
            if max_frames and out_count >= max_frames:
                break
        idx += 1
    cap.release()
    return frames, fps

# ---------------------------
# Face detection (Haar cascade fallback)
# ---------------------------
# Use cv2's frontalface haar cascade (commonly bundled)
CASCADE_PATH = cv2.data.haarcascades + "haarcascade_frontalface_default.xml"
face_cascade = None
if os.path.exists(CASCADE_PATH):
    face_cascade = cv2.CascadeClassifier(CASCADE_PATH)
else:
    face_cascade = None

def detect_face_boxes(frame, min_size=50):
    """
    Return list of (x,y,w,h) face boxes using Haar cascade fallback.
    If no faces found returns an empty list.
    """
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    boxes = []
    if face_cascade is not None:
        rects = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=4, minSize=(min_size, min_size))
        for (x,y,w,h) in rects:
            boxes.append((x,y,w,h))
    return boxes

def crop_and_resize(frame, box, size=(224,224)):
    x,y,w,h = box
    h_img, w_img = frame.shape[:2]
    # pad square around box a bit
    pad = int(0.25 * max(w,h))
    x0 = max(0, x-pad)
    y0 = max(0, y-pad)
    x1 = min(w_img, x+w+pad)
    y1 = min(h_img, y+h+pad)
    crop = frame[y0:y1, x0:x1]
    return cv2.resize(crop, size)

# ---------------------------
# Forensic features (heuristic mode)
# ---------------------------
def variance_of_laplacian(gray):
    return cv2.Laplacian(gray, cv2.CV_64F).var()

def high_freq_energy(img_gray):
    """
    Compute relative energy in high-frequency FFT coefficients.
    """
    f = fftpack.fft2(img_gray)
    fshift = fftpack.fftshift(f)
    magnitude = np.abs(fshift)
    # center low freq; compute fraction of energy outside central radius
    cy, cx = np.array(magnitude.shape)//2
    Y, X = np.ogrid[:magnitude.shape[0], :magnitude.shape[1]]
    radius = min(magnitude.shape)//8  # central low-freq radius
    mask_low = (X - cx)**2 + (Y - cy)**2 <= radius*radius
    total = magnitude.sum() + 1e-9
    high_energy = magnitude[~mask_low].sum()
    return high_energy / total

def block_artifact_score(img_gray, block=8):
    """
    Rough estimate of JPEG-like blocking artifacts by DCT block variance.
    Higher score -> more blockiness (often in recompressed / synthetic images).
    """
    h, w = img_gray.shape
    h_crop = (h // block) * block
    w_crop = (w // block) * block
    img = img_gray[:h_crop, :w_crop].astype(np.float32)
    # split into blocks and compute DCT-based variance
    s = 0.0
    cnt = 0
    for i in range(0, h_crop, block):
        for j in range(0, w_crop, block):
            blk = img[i:i+block, j:j+block]
            dct = cv2.dct(blk)
            # emphasize high-frequency coefficients (exclude DC)
            hf = dct.copy()
            hf[0,0] = 0.0
            s += np.mean(np.abs(hf))
            cnt += 1
    return s / (cnt + 1e-9)

def color_band_score(img, bins=16):
    """
    Detect color banding by looking for peaks in the histogram (low unique intensities).
    """
    lab = cv2.cvtColor(img, cv2.COLOR_BGR2LAB)
    L = lab[:,:,0].flatten()
    hist, _ = np.histogram(L, bins=bins, range=(0,255))
    # fraction of histogram mass in top 5 bins (indicates banding)
    top5 = np.sort(hist)[-5:].sum()
    return top5 / (len(L) + 1e-9)

def temporal_inconsistency(prev_gray, cur_gray):
    """
    Frame differencing normalized measure: high values indicate inconsistent motion.
    """
    diff = cv2.absdiff(prev_gray, cur_gray)
    return np.mean(diff) / 255.0

# Aggregate heuristic per-frame scoring
def heuristic_frame_score(frame, prev_frame_gray=None):
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    blur = variance_of_laplacian(gray)      # lower -> more blur (synth might be overly smooth)
    hf = high_freq_energy(gray)             # lower -> fewer high-freq details
    block_score = block_artifact_score(gray)
    band = color_band_score(frame)
    temp = 0.0
    if prev_frame_gray is not None:
        temp = temporal_inconsistency(prev_frame_gray, gray)
    # Normalize & combine heuristics into a single "fake-likelihood" score (0..1)
    # We invert blur and hf because small variance/high_freq -> real-like
    # These coefficients are empirical; for production you'd calibrate on labeled data.
    # All features are clipped and scaled to 0..1
    b_norm = 1.0 - np.clip(blur / 300.0, 0.0, 1.0)        # smaller blur => closer to 0
    hf_norm = 1.0 - np.clip(hf / 0.5, 0.0, 1.0)
    block_norm = np.clip(block_score / 10.0, 0.0, 1.0)
    band_norm = np.clip(band / 0.05, 0.0, 1.0)
    temp_norm = np.clip(temp / 0.03, 0.0, 1.0)
    # Weighted sum (weights chosen to emphasize unnatural frequency + temporal inconsistency)
    score = 0.28*b_norm + 0.28*hf_norm + 0.18*block_norm + 0.12*band_norm + 0.14*temp_norm
    return float(score), dict(blur=blur, hf=hf, block=block_score, band=band, temp=temp)

# ---------------------------
# Heuristic mode: analyze video
# ---------------------------
def analyze_video_heuristic(video_path, every_n_frames=5, max_frames=200):
    frames, fps = extract_frames(video_path, every_n_frames=every_n_frames, max_frames=max_frames)
    prev_gray = None
    scores = []
    per_frame_meta = []
    print(f"[heuristic] Extracted {len(frames)} frames (sample every {every_n_frames} frames).")
    for idx, frame in frames:
        score, meta = heuristic_frame_score(frame, prev_gray)
        prev_gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        scores.append(score)
        per_frame_meta.append((idx, score, meta))
    # aggregate
    mean_score = float(np.mean(scores)) if scores else 0.0
    median_score = float(np.median(scores)) if scores else 0.0
    std_score = float(np.std(scores)) if scores else 0.0
    # Simple thresholding -> label
    # Thresholds here are heuristic and should be tuned with labeled data
    if mean_score > 0.55:
        label = "Likely AI-generated / Manipulated"
    elif mean_score > 0.40:
        label = "Suspicious — possible synthetic content"
    else:
        label = "Likely Real / Natural"
    result = {
        "mean_score": mean_score,
        "median_score": median_score,
        "std_score": std_score,
        "label": label,
        "per_frame": per_frame_meta,
        "fps": fps
    }
    return result

# ---------------------------
# Deep-learning mode (train / predict)
# ---------------------------
def make_model(num_classes=2, backbone="resnet18", pretrained=True, device='cpu'):
    if not TORCH_AVAILABLE:
        raise RuntimeError("Torch not available; install torch & torchvision to use train/predict modes.")
    if backbone == "resnet18":
        model = models.resnet18(pretrained=pretrained)
        in_features = model.fc.in_features
        model.fc = nn.Linear(in_features, num_classes)
    else:
        raise NotImplementedError("Only resnet18 backbone implemented in this script.")
    return model.to(device)

def train_classifier(data_dir, save_ckpt, epochs=5, batch_size=16, lr=1e-4, device=None):
    """
    data_dir layout:
      data_dir/train/real/*.jpg
      data_dir/train/fake/*.jpg
      data_dir/val/real/*.jpg
      data_dir/val/fake/*.jpg
    Images should be face-crops.
    """
    if not TORCH_AVAILABLE:
        raise RuntimeError("Torch not available; install torch & torchvision to use train/predict modes.")
    device = device or ("cuda" if torch.cuda.is_available() else "cpu")
    print(f"[train] Using device: {device}")
    tfms = transforms.Compose([
        transforms.ToPILImage(),
        transforms.Resize((224,224)),
        transforms.RandomHorizontalFlip(),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485,0.456,0.406], std=[0.229,0.224,0.225])
    ])
    # Using ImageFolder for convenience
    train_dir = os.path.join(data_dir, "train")
    val_dir = os.path.join(data_dir, "val")
    train_ds = datasets.ImageFolder(train_dir, transform=tfms)
    val_ds = datasets.ImageFolder(val_dir, transform=transforms.Compose([
        transforms.ToPILImage(),
        transforms.Resize((224,224)),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485,0.456,0.406], std=[0.229,0.224,0.225])
    ]))
    train_loader = DataLoader(train_ds, batch_size=batch_size, shuffle=True, num_workers=2)
    val_loader = DataLoader(val_ds, batch_size=batch_size, shuffle=False, num_workers=2)
    model = make_model(num_classes=len(train_ds.classes), backbone="resnet18", pretrained=True, device=device)
    criterion = nn.CrossEntropyLoss()
    optimizer = torch.optim.Adam(model.parameters(), lr=lr)
    best_val_acc = 0.0
    for epoch in range(epochs):
        model.train()
        running_loss = 0.0
        running_correct = 0
        total = 0
        for imgs, labels in tqdm(train_loader, desc=f"Epoch[{epoch+1}/{epochs}]"):
            imgs = imgs.to(device)
            labels = labels.to(device)
            optimizer.zero_grad()
            outputs = model(imgs)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
            running_loss += float(loss.item()) * imgs.size(0)
            _, preds = torch.max(outputs, 1)
            running_correct += int((preds == labels).sum().item())
            total += imgs.size(0)
        train_loss = running_loss / total
        train_acc = running_correct / total
        # validation
        model.eval()
        val_correct = 0
        val_total = 0
        with torch.no_grad():
            for imgs, labels in val_loader:
                imgs = imgs.to(device)
                labels = labels.to(device)
                outputs = model(imgs)
                _, preds = torch.max(outputs, 1)
                val_correct += int((preds == labels).sum().item())
                val_total += imgs.size(0)
        val_acc = val_correct / (val_total + 1e-9)
        print(f"Epoch {epoch+1}: train_loss={train_loss:.4f}, train_acc={train_acc:.4f}, val_acc={val_acc:.4f}")
        # save best
        if val_acc > best_val_acc:
            best_val_acc = val_acc
            torch.save({
                "model_state": model.state_dict(),
                "classes": train_ds.classes
            }, save_ckpt)
            print(f"Saved best checkpoint to {save_ckpt} (val_acc={val_acc:.4f})")
    print("Training finished. Best val acc:", best_val_acc)

def predict_video_with_model(video_path, ckpt_path, every_n_frames=5, max_frames=300, device=None):
    """
    Uses the trained checkpoint to make per-frame predictions on face-crops and aggregates them.
    """
    if not TORCH_AVAILABLE:
        raise RuntimeError("Torch not available; install torch & torchvision to use train/predict modes.")
    device = device or ("cuda" if torch.cuda.is_available() else "cpu")
    checkpoint = torch.load(ckpt_path, map_location=device)
    classes = checkpoint.get("classes", ["real","fake"])
    model = make_model(num_classes=len(classes), backbone="resnet18", pretrained=False, device=device)
    model.load_state_dict(checkpoint["model_state"])
    model.eval()

    frames, fps = extract_frames(video_path, every_n_frames=every_n_frames, max_frames=max_frames)
    if len(frames) == 0:
        raise RuntimeError("No frames extracted.")
    # transform for inference
    tfm = transforms.Compose([
        transforms.ToPILImage(),
        transforms.Resize((224,224)),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485,0.456,0.406], std=[0.229,0.224,0.225])
    ])
    per_frame_preds = []
    for idx, frame in tqdm(frames, desc="Predicting frames"):
        boxes = detect_face_boxes(frame)
        if len(boxes) == 0:
            # fallback: use center crop of frame resized
            crop = cv2.resize(frame, (224,224))
            inp = tfm(crop).unsqueeze(0).to(device)
            with torch.no_grad():
                out = model(inp)
                probs = torch.softmax(out, dim=1).cpu().numpy()[0]
                pred_idx = int(np.argmax(probs))
                per_frame_preds.append((idx, classes[pred_idx], float(probs[pred_idx]), probs.tolist()))
        else:
            # pick largest face
            largest = max(boxes, key=lambda b: b[2]*b[3])
            crop = crop_and_resize(frame, largest, size=(224,224))
            inp = tfm(crop).unsqueeze(0).to(device)
            with torch.no_grad():
                out = model(inp)
                probs = torch.softmax(out, dim=1).cpu().numpy()[0]
                pred_idx = int(np.argmax(probs))
                per_frame_preds.append((idx, classes[pred_idx], float(probs[pred_idx]), probs.tolist()))
    # Aggregate: count frame votes, average probability for "fake" class if we know its index
    if "fake" in classes:
        fake_idx = classes.index("fake")
    else:
        fake_idx = 1 if len(classes) > 1 else 0
    fake_probs = np.array([p[3][fake_idx] for p in per_frame_preds])
    mean_fake_prob = float(fake_probs.mean()) if len(fake_probs)>0 else 0.0
    # decide threshold (0.5 default)
    label = "Likely AI-generated" if mean_fake_prob > 0.5 else "Likely Real"
    return {
        "mean_fake_probability": mean_fake_prob,
        "label": label,
        "per_frame": per_frame_preds,
        "fps": fps
    }

# ---------------------------
# CLI
# ---------------------------
def parse_args():
    p = argparse.ArgumentParser(description="AI-generated video detection (heuristic + DL)")
    p.add_argument("--mode", type=str, choices=["heuristic", "train", "predict"], required=True,
                   help="Mode: heuristic (no training), train (train classifier), predict (use checkpoint)")
    p.add_argument("--video", type=str, help="Video path for heuristic/predict modes")
    p.add_argument("--every", type=int, default=5, help="Sample every N frames")
    p.add_argument("--max_frames", type=int, default=300, help="Cap number of frames to process")
    p.add_argument("--data_dir", type=str, help="Data directory for training (train/val structure)")
    p.add_argument("--epochs", type=int, default=5, help="Training epochs")
    p.add_argument("--batch_size", type=int, default=16, help="Training batch size")
    p.add_argument("--save_ckpt", type=str, default="model.pth", help="Where to save checkpoint")
    p.add_argument("--ckpt", type=str, help="Checkpoint path for predict mode")
    return p.parse_args()

def main():
    args = parse_args()
    if args.mode == "heuristic":
        if not args.video:
            print("Provide --video for heuristic mode.")
            return
        res = analyze_video_heuristic(args.video, every_n_frames=args.every, max_frames=args.max_frames)
        print("=== Heuristic result ===")
        print(f"Label: {res['label']}")
        print(f"Mean score: {res['mean_score']:.4f}, Median: {res['median_score']:.4f}, Std: {res['std_score']:.4f}")
        print("Top 5 most suspicious frames (by score):")
        sorted_pf = sorted(res['per_frame'], key=lambda x: x[1], reverse=True)[:5]
        for idx, score, meta in sorted_pf:
            print(f" frame {idx}: score={score:.4f}, meta={meta}")
    elif args.mode == "train":
        if not args.data_dir:
            print("Provide --data_dir for training.")
            return
        train_classifier(args.data_dir, save_ckpt=args.save_ckpt, epochs=args.epochs, batch_size=args.batch_size)
    elif args.mode == "predict":
        if not args.video or not args.ckpt:
            print("Provide --video and --ckpt for predict mode.")
            return
        res = predict_video_with_model(args.video, args.ckpt, every_n_frames=args.every, max_frames=args.max_frames)
        print("=== Model prediction result ===")
        print(f"Label: {res['label']}")
        print(f"Mean fake probability: {res['mean_fake_probability']:.4f}")
        print("Sample per-frame predictions (first 10):")
        for item in res['per_frame'][:10]:
            print(item)
    else:
        print("Unknown mode.")

if __name__ == "__main__":
    main()


    
```
